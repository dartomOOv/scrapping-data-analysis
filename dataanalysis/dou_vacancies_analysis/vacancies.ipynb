```python
# Importing necessary libraries to work with vacancies data
import numpy as np
import pandas as pd

# Data retrieving using pandas and creating Series with expected technologies
data = pd.read_csv("../../vacancies.csv")
known_technologies = pd.Series([
    "Django", "DRF", "FastAPI", "PostgreSQL", "Vue", "Kubernetes", "Docker",
    "Redis", "MongoDB", "Azure", "RabbitMQ", "NodeJS", "TypeScript",
    "SQL", "Java", "Scala", "Python", "Flask", "Angular", "Prometheus", 
    "TensorFlow", "Pytest", "Elasticsearch", "Kafka", "Terraform", "Git",
    "GitHub", "AWS", "GraphQL", "MySQL", "JavaScript", "HTML", "CSS", 
    "SQLAlchemy", "React", "Vue3", "Selenium", "PySpark"
])

# Creating Dataframe with technologies and their number of mention in vacations
counter_mask = np.array([
    data["Description"].str.contains(technology).values
    for technology in known_technologies
])
tech_counts = counter_mask.sum(axis=1)
tech_data = pd.DataFrame({
    "Technology": known_technologies,
    "Counter": tech_counts
}).sort_values(by=["Counter", "Technology"], ascending=[True, True])

# Visualisation using Matplotlib
plot = tech_data.plot(kind="barh", figsize=(8, 10), legend=False, x="Technology", ylabel="Technologies", title="Technologies on Dou vacancies for Python developers.", color="cyan")
```

```python
# Save chart to the file
figure = plot.get_figure()
figure.savefig("charts/dou_vacancies.pdf")
```
